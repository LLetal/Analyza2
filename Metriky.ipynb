# Metrika a metrický prostor 11.1.1. (Def 1)
## Metrika
Mějme zobrazení $\rho: p \times p \to [0,\infty)$. Toto zobrazení nazýváme metrikou právě tehdy, když:
- 1. $\rho(x,y) \geq 0 \, \forall x,y$ a $\rho(x,y) = 0 <=> x=y$
- 2. $\rho(x,y) = \rho(y,x)$
- 3. $\rho(x,z) \leq \rho(x,y)+\rho$
## Metrický prostor
Množina, na níž je definovaná nějaká metrika $\rho$. Značí se $(P,\rho)$


# Norma a normovaný prostor 11.1.5 (Def 2)
Mějme V - vektorový prostor nad $\mathbb{R}$. Potom $||\cdot||: V \to [0,\infty)$ nazveme normou právě tehdy, když $\forall u,v$ a $\lambda \in \mathbb{R}$ platí:
- 1. $||u||\geq 0$
- 2. $||\lambda u||$ = $|\lambda|$ $||u||$
- 3. $||u+v|| \leq ||u||+||v||$

Každý takovýto prostor nazýváme $V$ nazýváme normovaným lineárním prostorem.

# Otevřená koule (V skriptách není (Def 4))

# Základní srandy s metrickým prostorem (Def 5)(11.3.13)
- 1. Množina $A \in P$ je otevřená právě tehdy, když $\forall x  \in A, \exists r >0: B(x,r)\subset A$
- 2. Množina $A$ je uzavřená právě tehdy, když $P$ bez $A$ je otevřená
- 3. Vnitřek množiny $A \subset P$ je $A^o =${$x \in A,\exists r >0,B(x,r)\subset A$}
- 4. Uzávěr množiny $A \subset P$ je definován jako $A^{-} = ${$x \in P,\forall x>0,B(x,r) \cap A \neq 0$}
- 5. Hranice množiny $A \subset P$ definujeme jako $jA = A^{-} \cap (P \setminus A)$

# KOnvergence v metrickém prostoru (Def 8)(11.2.1)
Nechť $(P,\rho)$ je metrický prostor a {$x_n$} $\subset P$ je posloupnost v něm. Řekneme, že $x_n$ konverguje k $x \in P$ pro $n \to \infty$ právě tehdy, když $\rho (x,x_n) \to 0$. Tehdy píšeme, že $x_n \to x$


# Ekvivalentní normy a metriky (Def 9)(11.2.6)
Nechť $(P,\rho_1),(P,\rho_2)$ jsou metrické prostory. Řekneme, že $\rho_1$ a $\rho_2$ jsou ekvivalentní právě tehdy, když $\forall x,y \in P$:
$$
c \rho_1(x,y) \leq \rho_2(x,y) \leq c_2\rho_1(x,y)
$$
Stejné tvrzení platí i pro normy.
$$
c ||x,y|| \leq ||x,y||\leq c_2 ||x,y||
$$

# Cauchyovská posloupnost (Definice 11) (11.5.1)
Nechť $(P,\rho)$ je metrický prostor. Řekneme, že posloupnost {$x_n$}$\subset P$ je cauchyovská v metrickém prostoru $(P,\rho)$, pokud:
$$
\forall \epsilon >0 \exists n_0 \in \mathbb{N} \, m,n > n_0 => \rho(x_m,x_n)<\epsilon
$$

# Úplný metrický prostor (Definice 11)(11.5.2)
O metrickém prostoru $(P,\rho)$ řekneme, že je úplný, pakliže každá cauchyovská posloupnost jeho prvků v něm konverguje.
## Bráňovy tipy a triky k tomuhle:
- Každá konvergentní posloupnost je vždy Cauchyovská
- Cauchyovská je konvergentní jen v úplném prostoru
- Ekvivalentní metriky/normy mají vždy stejné konvergentní a stejné Cauchyovské posloupnosti
- Ekvivalenstní metriky/normy jsou buď obě úplné, anebo obě neúplné 

# Pokrytí množiny (11.7.1)

Řekneme, že $\{ M_\alpha\}$ je pokrytí $A$, jestliže $A \subset \bigcup_{\alpha \in I}M_\alpha$. Jsou-li všechny množiny v systému $\{M_\alpha \}$ otevřené, hovoříme o otevřeném pokrytí.

# Separabilní prostor (Def 15)(11.4.4)

# Hustá podmnožina (Def 15) (11.4.1)

# Úplnost prostoru spojitých funkcí na uzavřeném intervalu v „nekonečno“ normě Věta 12 (11.6.3)
$(C([a,b]),||\cdot||_\infty)$ je úplný.
## Důkaz
Nechť {$f_n$}$\subset C([a,b])$ je cauchyovská posloupnost funkcí. Potom:
$$
\forall x \in [a,b] \,|f_n(x)-f_m (x)| \leq ||f_n - f_m||_\infty < \epsilon
$$
Ta nekonečná norma je ofc maximum, protože definice cauchyovské (nekonečná nám to posílá k nekonečnu a yk cauchyovskost ez). Vím, že:
$$
\lim_{m \to \infty} f_m(x) = f(x)
$$
A teď chceme tedy dokázat, že $f(x)$ je spojitá a že konverguje jako funkce. Konvergenci dokážeme dále:

$$
\forall m,n \geq n_0 \, \max_{t \in [a,b]} |f_n(t) - f_m(t)| < \epsilon
$$
V každém bodě mám nějakou hodnotu, vezmu z normy jejich maximum a položím to rovno $\epsilon$. Odtud jednoduše plyne:
$$
\epsilon + f_n (t) > f_m(t) > f_n(t)-\epsilon \, \forall t \in [a,b] 
$$
Limitní přechod k $m \to \infty$:
$$
\epsilon + f_n (t) \geq f(t) \geq f_n(t)-\epsilon \, \forall t \in [a,b] 
$$
Potom ale:
$$
\forall m,n \geq n_0 \, \max_{t \in [a,b]} |f_n(t) - f(t)| \leq \epsilon
$$
$f_n$ konverguje k $f$, jj, limita. Konverguje. Nyní dokažme spojitost.
\
V bodě $x_0 \in [a,b]$. Potom:
$$
\forall \epsilon >0 => \exists n_0 \in \mathbb{N} \, n \geq n_0 \, ||f_n - f||_\infty \leq \epsilon
$$
Víme, že jednotlivé funkce jsou spojité, ale nevíme, jestli celá limita:
$$
\exists \delta \, \forall x \in U_\delta (x_0) \cap [a,b] \, |f_n(x)-f_n(x_0)|< \epsilon
$$
To vede na trojúhelník:
$$
|f(x)-f(x_0)| \leq |f(x)-f_n(x)|+ |f_n(x)-f_n(x_0)| + |f(x_0)-f_n(x_0)| < 3\epsilon \, \, \text{Trojúhelníková nerovnost'}
$$

# Kompaktnost implikuje separabilitu (Věta 26)(11.6.10)
Pokud je $A \subset (P,\rho)$ kompaktní, tak musí být i separabilní.
## Důkaz
Pokud $A$ je kompaktní, potom platí:
$$
\forall \epsilon >0 \exists \text{konečná $\epsilon$-síť} 
$$
Lze to pokrýt konečným počtem otevřených koulí. To dokážeme sporem. Předpokládejme, že $\exists \epsilon > 0$ takové, že to nejde pokrýt. Vezměme tedy $x_1 \in A$. 
$$
A \notin B(x_1,\epsilon) =>x_2 \in A / B(x_1,\epsilon), \rho (x_1,x_2) \geq \epsilon
$$
No tak když přece neleží uvnitř té koule, která má poloměr $\epsilon$, tak mezi nima musí být vzdálenost větší, než $\epsilon$. Jedeme dál:
$$
A \notin \bigcup_{i = 1}^2 B(x_i,\epsilon) => x_3 \in A / \bigcup_{i = 1}^2 B(x_i,\epsilon) , \rho(x_3,x_1) \geq \epsilon \land \rho(x_3,x_2) \geq \epsilon  
$$
Extrapoluju na $n$ bodů. Posloupnost těchto bodů {$x_n$}$\in A$, $\rho(x_i,x_j) \geq \epsilon$ => není Cauchyovská a nejde z ní vybrat konvergentní podposloupnost, což je spor s kompaktostí a síto tedy musí existovat. 
\
Pokračujeme v důkazu. Označme si $\frac{1}{n}$ síť : $A_{\frac{1}{n}} = ${$x_1,\dots,x_m$}. Potom:
$$
K = \bigcup_{n = 1}^{\infty} A_{\frac{1}{n}} 
$$
je spočetné sjednocení spočetných množin. Současaně je hustá:
$$
\overline{K} \subset A \land K \subset A 
$$
A díky tomu, že $A$ je kompaktní, pak je i uzavřená a omezená. Současně:
$$
A \subset \overline{K} \text{protože} \forall x \in A \, \forall r > 0 \, B(x,r) \cap K \neq \emptyset
$$
$K$ nemůže být prázdný, protože jsme prokázali, že $A$ je všude pokryté koulemi.

# Cantorova věta o průniku kompaktů (Věta 27)(11.6.11)
Nechť K1 ⊃ K2 ⊃ . . . je posloupnost neprázdných kompaktů v metrickém prostoru. Pak T n∈N Kn je neprázdný kompakt. (kompakt = kompaktní množina)
## Důkaz
$\bigcap_{i=1}^\infty K_i$, všechny $K_i$ jsou kompaktní => jsou uzavřené => $\bigcap_{i=1}^\infty$ je též uzavřený. $\bigcap_{i=1}^\infty \subset K_1$ je kompaktní (ze znění věty). Když je něco podmnožinou kompaktní množiny, tak je též kompaktní. Vezměme $x_n \in K_n \subset K_1$. {$x_n$} $\subset$ $K_1$ kompaktní => $\exists {x_{n_k}} \to x \in K_1$. Podívám se na {$x_{n_2},\dots, x_{n_k}$}$\subset K_{n_2}$. To taky konverguje, protože se jenom přesouvám do další podmnožina, která je ale stále kompaktní. => $\exists${$x_{n_k} \to x \in K_{n_k}$} => $x\in \bigcap K_{n_k} = \bigcap K_{n_k}$. Čili každá množina má aspoň jeden prvek. $\square$ 


# Borelova pokrývací věta (Věta)(11.7.3)
Nechť $(P,\rho)$ je metrický prostor a $A \subset P$ je kompaktní. Pak lze z každého otevřeného pokrytí množiny $A$ vybrat konečné podpokrytí.
## Důkaz
A => kompaktní => A je separabilní =>Lindelölefa(z každého separabilního A můžeme vzít spočetnou)=> můžeme se omezit na spočetné pokrytí. 
\
Nechť {$M_j$}$_j ^\infty$ jsou spočetné otevřené pokrytí $A$. Definujeme si $G_j = \bigcup^j_{k = 1}M_k $, ukážeme, že $A \subset G_{n_0}$ pro nějaké $n_0$.
\
SPOREM. Předpokládejme, že to je teda lež. 
$$
K_n = A/ G_n \forall n \text{je neprázdné}
$$
$A_n$ je otevřená, ale potom musí $K_n$ nutně být uzavřená a zároveň podmnožina $A$. Protože $A$ je kompaktní, tak podmnožina $K_n$ je nutně taky kompaktní (viz předchozí věta) a neprázdná. ZÁroveň znovu z Cantora $K_1 $opačné subset. Samotný průnik $K_i$ je neprázdná, kompaktní => 
$$
\exists x \in \bigcap^\infty _{i=1} K_i = \bigcap^\infty_{i=1} (A/G_i) = A /\bigcup^\infty _{i=1}G_i = A/ \bigcup^\infty _{i=1}M_i
$$
=> $\exists x$mimo $\bigcup^\infty _{(i=1)} M_i$=> SPOR s pokrytím. $\square$.
