# Řada (Def 1)(9.1.1)
Nechť {$a_k$}$\subset  \mathbb{R}$ je posloupnost. Symbol $\sum_{k=1}^\infty a_k$ budeme označovat číselnou řadu. Pro $k \in \mathbb{N}$ se číslo $a_k$ nazývá k-tý člen, číslo $s_n = \sum_{k=1}^n a_k$ se nazývá n-tý částečný součet a {$s_n$} nazveme posloupností částečných součtů řady $\sum_{k=1}^\infty a_k$.


# Konvergence, divergence a oscilace řady (9.1.1)
Existuje.li vlastní limita $s = \lim_{n \to \infty}s_n$ říkáme, že řada konverguje. Pokud je uvedená limita nevlastní, řada diverguje a pokud limita částečných součtů neexisutje, tak řada osciluje. V prvních dvou případech nazveme $s$ součtem řady. Píšeme:
$$
\sum_{k=1}^n a_k = s
$$


# Absolutní konvergence (Def 11)(9.1.15)
Říkáme, že číselná řada $\sum_{k=1}^\infty a_k$ konverguje absolutně, jestliže konverguje $\sum_{k=1}^\infty |a_k|$. Říkáme, že řada $\sum_{k=1}^\infty a_k$ konverguje neabsolutně, jestliže konverguje $\sum_{k=1}^\infty a_k$, ale nekonverguje $\sum_{k=1}^\infty |a_k|$

# Přerovnávání řady (Def 14) (9.4.1)
Nechť {$a_k$}$\subset \mathbb{R}$ a $\phi : \mathbb{N} \to \mathbb{N}$ je bijekce. Pak řadu $\sum_{k=1}^\infty a_{\phi(a_k)}$ nazveme přerovnáváním řady $\sum_{k=1}^\infty a_k$.  

# Exp, sin, cos (10.4.2)
$$
e^z = \sum_{k=0}^\infty \frac{z^k}{k!}
$$
$$
\sin(z) = \sum_{k=0}^\infty (-1)^k \frac{z^{2k +1}}{(2k +1)!}
$$
$$
\cos(z) =\sum_{k=0}^\infty (-1)^k \frac{z^{2k}}{(2k)!}
$$

# Cauchyho součin (Def)(9.4.12)
Mějme dvě číselné posloupnosti {$a_n$}$_{n=1}^\infty$, {$b_n$}$_{n=1}^\infty$ tak, že $\sum_{n=1}^\infty a_n$, $\sum_{n=1}^\infty b_n$ konverguje absolutně. Potom $\sum_{i,j=1}^\infty a_i b_j$ je Cauchyho součin.

# Mocninná řada (Def bez čísla)(10.1.1)
Nechť {$a_n$}$\subset \mathbb{C}$ a $z_0 \in \mathbb{C}$. Pak řadu
$$
\sum_{k=0}^\infty a_k(z-z_0)^k
$$
nazáváme mocninnou řadou se středem $z_0$. Čísla $a_k$ nazýváme koeficienty mocninné řády.

# Nutná podmínka konvergence (Věta 2) (9.1.8)
Nechť řada $\sum_{n=1}^\infty a_n$ konverguje. Potom musí platit, že $\lim_{k \to \infty} a_k =0$
## Důkaz
$$
\lim_{k \to \infty} a_k = \lim_{k \to \infty} (S_k - S_{k-1}) = \lim_{k \to \infty} S_k - \lim_{k\to \infty} S_{k-1} = L-L =0
$$
(L je nějaký částečný součet, ke kterému to konverguje a tím, a využíváme toho, že když to konverguje k u $k$-tého členu, tak u $k-1$ u nekonečna už to tam de facto bude také).

# Srovnávací kritérium (Věta 4)(9.1.19)

## Důkaz
$$
0 \leq \sum_{k=1}^n a_k \leq \sum_{k=1}^n b_k
$$
$$
\sup_{n \in \mathbb{N}}\sum_{k=1}^n a_k \leq \sup_{n \in \mathbb{N}} \sum_{k=1}^n b_k
$$
$$
A \leq B
$$
pozn.: Hovoříme zde o supremech částečných součtů. Alternativní přes B-C podmínku:
\\

# Podílové kritérium (Věta 5)(9.2.1)
Nechť {$a_k$}, {$b_k$}$\subset [0,\infty)$, $k_0 \in \mathbb{N}$ a je splněna alespoň jedna z podmínek:
- I) $a_k \geq b_k$ $\forall k \geq k_0$
- II) $\frac{a_{k+1}}{a_k} \geq \frac{b_{k+1}}{b_{k}} \forall k\geq k_0$
a k tomu všemu $\sum_{k=1}^\infty a_k$ konverguje. 

Potom $\sum_{k=1}^\infty b_k$ rovněž konverguje.
## Důkaz
U obou podmínek můžeme předpokládat, že $k_1$ jinak vhodným způsobem převedem $k_0 -1$ prvních členů zkoumaných řad, protože můžeme prostě. Platí-li podmínka I, výsledek plyne ze srovnávacího kritéria (9.1.19), které můžeme zkrátka užít z podmínek. Platí-li podmínka II, potom pro všechna $k \in \mathbb{N}$ máme:
$$
a_k = \frac{a_k}{a_{k-1}} \frac{a_{k-1}}{a_{k-2}} \dots \frac{a_2}{a_1}a_1 \geq \frac{b_k}{b_{k-1}} \frac{b_{k-1}}{b_{k-2}} \dots \frac{b_2}{b_1}b_1 = \frac{a_1}{b_1}b_k
$$
Odtud $b_k \leq \frac{b_1}{a_1}a_k$ pro všechna $k \in \mathbb{N}$. Napravo máme členy konvergentní řady, proto lze znovu užít srovnávací kritérium a je vybaveno. $\frac{b_1}{a_1}a_k$ je jen nějaké nulové číslo krát nenulové, což ve výsledku dává z podmínky o konvergenci smysl.

# Limitní srovnávací kritérium (Věta 8)(9.2.3)
Nechť {$a_k$}, {$b_k$}$\subset (0,\infty)$ a dále nechť $\lim_{k \to \infty} \frac{a_k}{b_k} \in (0,\infty)$. Pak $\sum_{k=1}^\infty a_k$ konverguje právě tehdy, když $\sum_{k=1}^\infty b_k$ konverguje.

\
Jestliže {$a_k$}, {$b_k$}$\subset (0,\infty)$ a $\lim_{k \to \infty} \frac{a_k}{b_k} = 0$ Pak:
-  I) Když$\sum_{k=1}^\infty b_k$ konverguje, pak $\sum_{k=1}^\infty a_k $ konverguje. 
- II) Když $\sum_{k=1}^\infty a_n diverguje$, pak $\sum_{k=1}^{\infty} b_n$ taktéž diverguje.

## Důkaz
Nejprve dokažme první řást kritéria. Označme si $L = \lim_{k\to \infty}\frac{a_k}{b_k}$. Z definice limity potom existuje takové $k_0 \in \mathbb{N}$ takové, že:
$$
\frac{L}{2}b_k \leq a_k \leq 2L b_k \, \text{pro $k\geq k_0$}
$$
Nyní použejeme podílové kritérium (9.2.1) vedoucí na srovnávací (9.2.3). Druhou část dokazujeme stejné, přičemž $\lim_{k\to\infty} \frac{a_n}{b_n}=0$ $\forall \epsilon >0 \, \exists n_0 \geq 1:$ $n\geq n_0$ $\frac{a_n}{b_n}\leq \epsilon$, což vede zase na srovnávací kritérium. Radši se zeptat

# Integrální kritérium (Věta 10)(9.2.5)
Nechť $a\in \mathbb{N}$ a $f: \mathbb{R} \to \mathbb{R}$ je spojitá, kladná a nerostoucí na $[a,\infty]$. Pak:
$$
\sum_{k=a}^\infty f(k) \text{konverguje} <=> (N)\int_a ^\infty f dx \in \mathbb{R}
$$
## Důkaz
Díky monotonii funkce $f$ máme (protože očekáváme, že konverguje, tak taky očekáváme, že její hodnota bude klesat):
$$
f(k+1) \leq (N) \int_{k}^{k+1} f dx \leq f(k)
$$
Proto pro libovolné $n \in \mathbb{N}$ $n>a$, platí:
$$
\sum_{k = a+1}^{n+1} f(k) = \sum_{k=a}^n f(k+1) \leq (N) \int_a^{n+1} f dx\leq \sum_{k=a}^n f(k)
$$
Pokud Newtonův integrál konverguje, je (neklesající) primitivní funkce omezená a nutně pak jsou podle leé části našeho odhadu omezené (neklesající) částečně součty řady $\sum_{k=a}^\infty f(k)$. Tato řada proto konverguje.

# O přerovnání absolutně konvergentní řady (Věta 15) (9.4.6)
Nechť {$a_k$}$\subset \mathbb{R}$ a řada $\sum_{k=1}^\infty a_k$ konverguje absolutně. Pak každé její přerovnání konverguje absolutně a má stejný součet.
## Důkaz
Mějme $\sum_{k=1}^\infty b_k$ přerovnání původní řady. Nejprve uvažme jednoduchý případ, kdy jsou všechny členy {$a_k$} nezáporné. Je-li $n \in \mathbb{N}$, pak existuje $k_0 \in \mathbb{N}$ takvé, že {$b_1, \dots ,b_n$}$\subset ${$a_1, \dots ,a_{k_0}$}, a proto
$$
\sum_{k=1}^{n}b_k \leq \sum_{k=1}^{k_0} a_k \leq \sum_{k=1}^\infty a_k \in \mathbb{R}
$$
Odtud plyne, že $\sum_{k=1}^\infty b_k$ má omezené monotonní částečné součty, tedy absolutně konverguje a rovněž platí, že$\sum_{k=1}^\infty b_k \leq \sum_{k=1}^\infty a_k$. Prohozením rolí {$a_k$} a {$b_k$} dostáváme obrácenou nerovnost. Proto jsou součty obou řad stejné.
\

V obecném příapdě si napíšeme, že $\sum_{k=1}^\infty b_k = \sum_{k=1}^\infty b_k^+ - \sum_{k=1}^\infty b_k ^-$. Por každou z řad na pravé straně platí výsledek dokázaný výše. Odtud
$$
\sum_{k=1}^\infty b_k = \sum_{k=1}^\infty b_k^+ - \sum_{k=1}^\infty b_k^- = \sum_{k=1}^\infty a_k^+ - \sum_{k=1}^\infty a_k^- = \sum_{k=1}^\infty a_k
$$
Abslutní konvergence potom plyne z konvergence kladných a záporných částí

# Cachyho věta o součinu řad (Věta nemá číslo)(9.4.12)
Mějme dvě číselné posloupnosti {$a_n$}$_{n=1}^\infty$, {$b_n$}$_{n=1}^\infty$ tak, že $\sum_{n=1}^\infty a_n$, $\sum_{n=1}^\infty b_n$ konvergují absolutně. Potom $\sum_{i,j=1}^\infty a_i b_j$ je Cauchyho součin.
$$
\sum_{i,j =1} a_i b_j = (\sum_{k=1}^\infty a_k)(\sum_{k=1}^\infty b_k).
$$
## Důkaz
Definujme $\tilde{A}_n = \sum_{k=1}^n |a_k|$,$\tilde{B}_n = \sum_{k=1}^n |b_k|$ pro každé $n \in \mathbb{N}$. Dále definujme bijekci $\phi: \mathbb{N}\to \mathbb{N^2}$(jednotlivé složky budeme později značit $\phi_1$,$\phi_2$) zaveďme si konstrucki:
$$
(1,1),(2,1),(2,2),(1,2),(3,1),(3,2),(3,3),(2,3),(1,3)\dots
$$
Dále definujme $S_n = \sum_{k=1}^n a_{\phi_1(k)} b_{\phi_2(k)}$ a $\tilde{S}_n = \sum_{k=1}^n |a_{\phi_1(k)}| |b_{\phi_2(k)}|$ pro každé $n \in \mathbb{N}$. Absolutní konvergence Cauchyho součinu řad plyne z toho, že {$\tilde{S}_n$} je neklesající posloupnost splňující:
$$
\tilde{S}_{n^2} = \sum_{k=1}^{n^2} |a_{\phi_1(k)}| |b_{\phi_2(k)}| = \sum_{i=1}^n a_{i} \sum_{i=1}^n b_{j} = \tilde{A}_n \tilde{B}_n \leq \sum_{i=1}^\infty |a_{i}| \sum_{j=1}^\infty |b_{j}|
$$
Proto je také $\sum_{i=1}^\infty a_{\phi_1 (k)}b_{\phi_2 (k)}$ konvergentní a platí prvotní rovnost.

# Derivace mocninné řady (Věta 22)(10.1.9)
Nechť {$a_k$}$\subset \mathbb{C}$. Pak pro $x \in (-R,R)$, kde $R \geq 0$ je poloměr konvergence příslušné mocninné řady, platí:
$$
(\sum_{k=0}^\infty a_k x^k)' = \sum_{k=1}^\infty k a_k x^{k-1}
$$
## Důkaz
Provedem přímé ověření definice derivace. Přerpokládejme $R>0$, jinak není co dokazovat. Zafixujme $x \in (-R,R)$ a $\delta  <R-|x|$. Pak pro $h \in (-\delta, \delta) \backslash \{0\}$ má podle aritmetiky konvergentních řad smysl nějaká veličina:
$$
\Psi(x,h) = \frac{\sum_{k=0}^\infty a_k(x+h)^k - \sum_{k=0}^\infty a_k x^k}{h} - \sum_{k=1}^\infty k a_k x^{k-1} = \sum_{k=2}^\infty (a_k \frac{(x+h)^k - x^k}{h} - ka_k x^{k-1}).
$$
(Přesunutí můžeme provést proto, že pro $i=0$ a $i=1$ je výraz nulový) Naším cílem je dokázat, že $\Psi(x,h) \to 0$ pro $h \to 0$. Použijeme hrůzostrašné Lemma 10.1.8 a skutečnost, že $\sum_{k=2}^\infty k(k+1)|a_k|(|x|+\delta)^{k-2}$ konverguje (neboť $|x|+\delta<R$), potom z Lemma 10.1.7. Pak:
$$
|\Psi(x,h)| = \frac{1}{|h|}|\sum_{k=2}^\infty a_k((x+h)^k - x^k - kx^{k-1}h)|
$$
$h$ vytýkáme proto, že to je konstanta. Pokračujme dál:
$$
= \frac{1}{|h|}|\sum_{k=2}^\infty a_k(\binom{k}{2} x^{k-2}h^2 + \dots + \binom{k}{k}h^k)|
$$
Vykrátí se mi první členy, protože se ti to pokrátí z binomické věty.
$$
\leq \frac{1}{|h|}\sum_{k=2}^\infty |a_k|(\binom{k}{2}|x|^{k-2}|h|^2 + \dots + \binom{k}{k}|h|^k) = \frac{1}{|h|}\sum_{k=2}^\infty |a_k|((|x|+|h|)^k - |x|^k - k|x|^{k-1}|h|) \leq \frac{1}{|h|}\sum_{k=2}^\infty |a_k|\frac{k(k-1)(|x|+|h|)^{k-2}}{2}|h|^2 \leq \frac{1}{2}|h|\sum_{k=2}^{\infty}|a_k|k(k-1)(|x|+\delta)^{k-2} = C|h| \to 0. \square 
$$

# Abelova věta (Věta 25)(10.2.8)
Mějme komplexní posloupnost {$a_n$}$^\infty _{n=0}$ a příslušná mocninná řada $f(z)= \sum_{k=0}^\infty a_k z^k$ má poloměr konvergence $R \in (0,\infty)$. Je-li $\phi \in [0,2\pi)$ takové, že pro $z = R e^{i\phi}$ konverguje řada $\sum_{k=0}^\infty a_k z^k$, pak $t \rightarrow f(te^{i\phi})$ je spojitá na $[0,R]$ i na poloměru_.
## Důkaz
Bez důkazu

# O konvergenci mocninné řady (Bez bráni čísla)(10.1.2)
$$
R = \frac{1}{\lim \sup_{k \to \infty}\sqrt[k]{|a_k|}} \, \text{s konvencí $\frac{1}{0} = \infty$ a $\frac{1}{\infty}=0$}
$$
\
Pak:
- I) Řada $\sum_{k=0}^\infty a_k z^k$ konverguje absolutně na $B_R$
- II) Řada $\sum_{k=0}^\infty a_k z^k$ nekonverguje na {$z \in \mathbb{C}: |z|>R$}
- III) Existuje-li $\lim_{k \to \infty} |\frac{a_k}{a_{k+1}}|$, pak se rovná $R$.
- IV) Existuje-li $\lim_{k \to \infty} \sqrt[k]{|a_k|}$, pak se rovná $\frac{1}{R}$

## Důkaz
Poslední část IV je zřejmá přímo z prvníh ořádku. Část I plyne z odmocninového kritéria. Dokažme si tedy II pro případ $R \in (0,\infty)$. V tomto případě pišme $|z|=(1+ \delta )R$, kde $\delta >0$, a máme
$$
|a_k z^k| = |a_k||z|^k = ((1+\delta) \frac{\sqrt[k]{|a_k|}}{\lim \sup_{n\to \infty}\sqrt[n]{|a_n|}})^k
$$
Z definice veličiny $\lim \sup$ musí být vnitřek závorky větší nž jedna pro nekonečně mnoho, což vede na porušení nutné podmínky konvergence (srovnání s geometrickou řadou). Pro $R = \infty$ není moc co dokazovat, stejně tak pro $R=0$.
\
Dokažme si tedy zbývající část III. Položme $r = \lim_{k \to \infty}|\frac{a_k}{a_{k+1}}|$. Pro zafixované $z\in \mathbb{C}$ máme:
$$
\frac{|a_{k+1}z^{k+1}|}{|a_k z^k|} = \frac{|z|}{|\frac{a_k}{a_{k+1}}|} \rightarrow^{k \to \infty} \frac{|z|}{r}.
$$
Dle podílového kritéria naše řada konverguje absolutně pro $|z|<r$, naopak pro $|z|>r$ absolutně konvergoat nemůže. Podle výsledků výše pak $r = R$
